```python
import networkx as nx 
import matplotlib.pyplot as plt 


def create_cycle(num):
    network = nx.Graph() 
    nodes = range(num)
    network.add_nodes_from(nodes) 
    nx.add_cycle(network, nx.nodes(network))
    nx.draw_networkx(network, with_labels=True)
    return network
```
